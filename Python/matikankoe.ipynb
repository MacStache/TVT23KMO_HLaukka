```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# Mritelln aika-akseli
t = np.linspace(0, 2 * np.pi, 1000)  # Aikavli ja pisteiden mr
dt = t[1] - t[0]  # Aika-askel

# Mritelln signaali
signal = np.sin(12 * t) + np.sin(18 * t) + np.sin(30 * t) + np.sin(48 * t) + np.sin(91 * t)

# Piirretn signaali (ilman tai kohinalla)
plt.figure(figsize=(10, 6))
plt.plot(t, signal)
plt.title('Signaali useilla siniaalloilla')
plt.xlabel('Aika (t)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()

# Tehospektrin laskenta (Fourier-muunnos)
N = len(t)  # Nytteiden mr
yf = fft(signal)  # Fourier-muunnos signaalista
xf = fftfreq(N, dt)[:N//2]  # Taajuusakseli

# Piirretn tehospektri
plt.figure(figsize=(10, 6))
plt.plot(xf, 2.0/N * np.abs(yf[:N//2]), label='Tehospektri')
plt.title('Signaalin tehospektri (Fourier-muunnos)')
plt.xlabel('Taajuus (Hz)')
plt.ylabel('Teho')
plt.grid(True)
plt.axis([0,10,0,5])
plt.legend()
plt.show()
```
