```python
import numpy as np
import matplotlib.pyplot as plt

# Määritellään x-arvot
x = np.linspace(-0.51 * np.pi, 0.51 * np.pi, 400)

# Määritellään funktio f(x) = 3*sin(2x + 1)
y = 3/2 * np.sin(2 * x + 1)

# Piirretään kuvaaja
plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$f(x) = 3\sin(2x + 1)$', color='blue')

# Asetetaan otsikot ja ruudukko
plt.title(r'Kuvaaja $f(x) = 3\sin(2x + 1)$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)

# Näytetään selite
plt.legend()

# Näytetään kuvaaja
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

t= np.arange(0,10,0.01)
f= 2.5*np.sin(3.2*t)
plt.figure(figsize=(12,4))
plt.plot(t,f)
# Näytetään selite
plt.legend()
plt.grid()
# Näytetään kuvaaja
plt.show()
```

```python
# Määritellään aikaväli
t = np.arange(0, 20, 0.01)

# Muutetaan kulmataajuus
omega = 2 * np.pi

# Määritellään funktio
f = 2.5 * np.sin(omega * t)

# Piirretään kuvaaja
plt.figure(figsize=(12, 4))
plt.plot(t, f)
plt.title("Sinimuotoinen funktio, jakson aika 0.5 sekuntia")
plt.xlabel("Aika (s)")
plt.ylabel("Amplitude")
plt.grid()
plt.legend(["$2.5 \sin(4 \pi t)$"])
plt.show()
```
