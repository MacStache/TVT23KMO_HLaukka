```python
import numpy as np
import matplotlib.pyplot as plt

# Mritetn aikavektori ja signaali
dt = 0.01  # Nytteenottovli
t = np.arange(0, 1, dt)  # Aikavektori 0:sta 1:een sekuntiin
F = np.sin(2 * np.pi * t) + np.cos(2 * np.pi * t) + np.cos(6 * np.pi * t)

# Piirretn alkuperinen signaali
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.plot(t, F)
plt.title('Alkuperinen signaali')
plt.xlabel('Aika (s)')
plt.ylabel('Amplitudi')

# Lasketaan signaalin FFT
fft_values = np.fft.fft(F)
n = len(F)  # Datapisteiden lukumr signaalissa
fs = 1 / dt  # Nytteenottotaajuus
fft_freq = np.fft.fftfreq(n, d=dt)

# Lasketaan tehospektri (PSD)
psd = (np.abs(fft_values) ** 2) / n

# Piirretn FFT suuruus
plt.subplot(1, 3, 2)
plt.plot(fft_freq, np.abs(fft_values))
plt.title('Signaalin FFT')
plt.xlabel('Taajuus (Hz)')
plt.ylabel('Suuruus')
plt.axis([1,10,0,70])
plt.grid(True)
plt.xlim(0, 10)  # Rajataan X-akseli vlille 0-10 Hz

# Piirretn tehospektri (PSD)
plt.subplot(1, 3, 3)
plt.plot(fft_freq, psd)
plt.title('Signaalin tehospektri (PSD)')
plt.xlabel('Taajuus (Hz)')
plt.ylabel('Teho')
plt.axis([1,10,0,70])
plt.grid(True)
plt.xlim(0, 10)  # Rajataan X-akseli vlille 0-10 Hz

plt.tight_layout()
plt.show()
```
